# Evaluación de una función de manera paralela

```python
import numpy as np
import random
import math
from multiprocess import Pool
```

```python
# variables
lim_inf = 0
lim_sup = 500
cant_valores = 10000
num_procesadores = 15
```

```python
def nivelacion_cargas(valores, procesadores):
     mod = len(valores) % procesadores  # por si sobran valores
     ind = math.floor(len(valores)/procesadores )   # para saber cuantos le toca a cada uno
     final = []    # para guardar las particiones

     # inicializar los limites inferiores y superiores
     li = 0
     ls = ind

     for i in range(procesadores):      # para un tiempo O(procesadores)
          ls = li+ind
          if i < mod:    # para repartir los valores que sobran
               ls += 1
          final.append(valores[li:ls])
          li = ls
     return final
```

```python
def aleatorios_rango(inf,sup,cant_valores):
     lista = np.random.randint(inf,sup,cant_valores)
     return lista
```

```python
lista = aleatorios_rango(lim_inf, lim_sup, cant_valores)
print("Lista --> ", lista)

# balanceo de cargas
cargas = nivelacion_cargas(lista, num_procesadores)

# función lambda
funcion = lambda x: x ** 2

res = []
for i in range(len(cargas)):
     with Pool(num_procesadores) as pool:
          # Se guardan los resultados en una variable
          res.append(pool.map(funcion, cargas[i]))

          resultado = [item for mini_lista in res for item in mini_lista]

print("Resultado --> ", resultado)
```
